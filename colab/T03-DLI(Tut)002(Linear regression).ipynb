# What is going on:
* You are a data scientist analyzing the relationship between years of experience and salary for a company. Your task is to build a model to predict the salary based on years of experience using a simple linear regression model.
* Explanation: Simple linear regression involves one dependent **(salary)** and one independent **(years of experience)** variable. The goal is to fit a straight line that best represents the relationship between these variables.

# Simple Linear Regression Model

*   Involves two variables: one dependent and one independent, represented in a two-dimensional space.
*   Predicts dependent variable values based on the independent variable using a straight line function.
*   Aims to accurately represent the relationship between variables.

Model Evaluation

*   Assessing predictions by comparing them to test data
points.
*   Acknowledges that while some predictions may be close, others may deviate from the line.
*   Considered the simplest regression model, with more complex models discussed in subsequent articles.

Improving Simple Linear Regression

* Utilizes the Scikit-Learn module for model building.
* Emphasizes the importance of data preprocessing before model fitting.
* Mentions the use of a random state parameter for result reproducibility.

Assumptions for Simple Linear Regression
* Data points should follow a normal distribution and exhibit equal variance.
* Non-compliance with these assumptions may necessitate the use of nonparametric tests, especially in scenarios like stock price data collection over time.

Importing the essential libraries

```python
#Simple Linear Regression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Read the dataset and split the data into independent values and target values

```python
# Mount your Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Importing the dataset
dataset = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Salary_Data.csv')
X = dataset.iloc[:, 0:1].values
y = dataset.iloc[:, 1].values
```

Split the value into training and testing

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

Train the Linear Regression with training dataset

```python
# Fitting Simple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Predict the model with test dataset

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

Visualising the Training set results

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'green')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
