**1- Data Import and Preprocessing:**

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**2- Load Dataset:**

```
# Mount your Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Importing the dataset
dataset = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Social_Network_Ads.csv')
dataset.head()
```

```
X = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, 4].values
```

**3- Split Data:**

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```
X_train[:10]
# This line displays the first 10 rows of the X_train NumPy array.
```

**4- Feature Scaling:**

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
X_train[:10]
```

**5- Train Logistic Regression Model:**

This code snippet is where the actual machine learning model is created and trained.

**Importing the Model:**

from sklearn.linear_model import LogisticRegression: This line imports the LogisticRegression class from the sklearn.linear_model module. This class provides the necessary tools to build a Logistic Regression model.

**Creating the Model:**

classifier = LogisticRegression(random_state = 0): Here, we create an instance of the LogisticRegression class and assign it to the variable classifier.
random_state = 0: This parameter ensures that the results are reproducible. By setting it to a fixed value (0 in this case), the model will be initialized the same way every time the code is run, leading to consistent outcomes.

**Training the Model:**

classifier.fit(X_train, y_train): This line is where the actual training happens.
The fit method of the LogisticRegression object is called, taking the training data (X_train and y_train) as input.
X_train: Represents the features (independent variables) of the training dataset.
y_train: Represents the target variable (dependent variable) of the training dataset.
During the training process, the model learns the relationships between the features and the target variable by adjusting its internal parameters to minimize prediction errors on the training data.



```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

**6- Predict Test Set Results:**

This line is where the trained Logistic Regression model is used to make predictions on the test dataset.

```
y_pred = classifier.predict(X_test)
```

**7- Confusion Matrix and Accuracy:**

1. Import necessary libraries for confusion matrix calculation and visualization.
2. Calculate the confusion matrix using the true labels and predicted labels.
3. Define labels for the categories being predicted (e.g., 'No', 'Yes').
4. Create a figure and a subplot to display the confusion matrix.
5. Display the confusion matrix as a color-coded image within the subplot.
6. Add a colorbar to the plot to provide a scale for the colors.
7. Set labels for the x and y axes of the plot, indicating predicted and true values.
8. Display the created plot showing the visualized confusion matrix.




```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

labels = ['No', 'Yes']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

**8- Visualization (Training and Test Set):**

1. Import Necessary Libraries: Import ListedColormap for creating colormaps.
2. Prepare Data: Assign the training data (features and target variable) to variables for easier use.
3. Create a Grid: Generate a grid of points covering the range of feature values present in the training data. This grid will be used to visualize the decision boundary of the model.
4. Plot the Decision Boundary: Use the trained Logistic Regression model to predict the class label for each point on the grid. Then, create a filled contour plot using these predictions, coloring the regions of the grid based on the predicted classes.
5. Set Plot Limits: Adjust the limits of the x and y axes to ensure the entire grid is visible in the plot.
6. Plot Training Data Points: Plot the actual data points from the training set onto the grid, using different colors to represent the different classes.
7. Add Labels and Title: Add a title, axis labels (for "Age" and "Estimated Salary"), and a legend to the plot to provide context and clarity.
8. Display the Plot: Show the created plot to visualize the decision boundary and the training data.

```
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```
# Visualising the Test set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

**9- Save Model:**
Save the trained model to a file using pickle.

```
import pickle
with open('/content/drive/My Drive/Colab Notebooks/regressor.pkl', 'wb') as f:
    pickle.dump(classifier, f)
```

**10- Load Model and Make Prediction:**

```
loaded_model = pickle.load(open('/content/drive/My Drive/Colab Notebooks/regressor.pkl', 'rb'))

# Get user input for age and salary
age = float(input("Enter Age: "))
salary = float(input("Enter Estimated Salary: "))

# Create data for prediction
data = [[age, salary]]

# Scale the data using the same StandardScaler object (sc)
data = sc.transform(data)

# Make the prediction
y_pred = loaded_model.predict(data)

# Interpret and print the prediction
if y_pred[0] == 1:
  print('Buy')
else:
  print('Not Buy')
```
